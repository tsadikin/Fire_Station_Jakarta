```python
df = pd.read_excel('Data_Kebakaran_Compiled.xlsx')
df
```

```python
df = df[df.tanggal != 2015]
df
```

```python
df['tanggal'] = pd.to_datetime(df['tanggal'])
df['day'] = df['tanggal'].dt.day_of_week
df['day'] = df['day'] + 1
df
```

```python
#shows distribution of each day
ax = sns.distplot(df['day'],
                  bins=50,
                  kde=True,
                  color='green',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
 
plt.show()
```

```python
#fit norma ldist
time = []
plt.hist(df['day'], bins=100, density=True)
[loc_fit, scale_fit]=scipy.stats.norm.fit(df['day'])
print(loc_fit)
print(scale_fit)
x = np.linspace(np.min(df['day']), np.max(df['day']))

plt.plot(x, scipy.stats.norm.pdf(x, loc=loc_fit, scale=scale_fit),)

plt.xlabel("speed")

plt.ylabel("Prob Dense")
plt.show()
r = sp.stats.norm.rvs(loc=[loc_fit, scale_fit][0], scale=[loc_fit, scale_fit][1],size=(1000000))
time.append(r)
```

```python
time
```

```python
timedf = pd.DataFrame(np.concatenate(time))
timedf = timedf.rename(columns={0: 'Time'})
timedf
```

```python
#create data frame for coordinates of demand simulation and day of happening
alldf['day'] = timedf['Time']
alldf['day'] = alldf['day'].round()
alldf
```

```python
#using uniform distribution to generate time column
s = np.random.uniform(1,24,1000000)
s
```

```python
count, bins, ignored = plt.hist(s, 15, density=True)
plt.show()
```

```python
clock = pd.DataFrame(s)
clock[0]
```

```python
import datetime
#hours
clock['Hours'] = clock[0]//1
clock
```

```python
alldf['Hours'] = clock['Hours']
alldf['Time'] = alldf['day'] + alldf['Hours']/24
alldf
```

```python
#The time schedule for each mdoel
#1-2.22 = 2, 2.22-3.88 =4, 3.88-5.48=3, 5.48-6.96=0, 6.96-8=1
Cluster = []
for i in alldf['Time']:
    if 1 < i and i <= 2.22:
        x = 2
    elif 2.22<i and i <=3.88:
        x = 4
    elif 3.88<i and i <=5.48:
        x = 3
    elif 5.48<i and i <=6.96:
        x = 0
    elif 6.96<i and i<=8:
        x = 1
    Cluster.append(x) 
```

```python
Cluster
```

```python
dfCluster = pd.DataFrame(Cluster)
dfCluster
```

```python
alldf['Cluster'] = dfCluster[0]
alldf
```

```python
data0 = alldf[alldf.Cluster==0]
data1 = alldf[alldf.Cluster==1]
data2 = alldf[alldf.Cluster==2]
data3 = alldf[alldf.Cluster==3]
data4 = alldf[alldf.Cluster==4]
```

```python
data0 = data0.reset_index()
data1 = data1.reset_index()
data2 = data2.reset_index()
data3 = data3.reset_index()
data4 = data4.reset_index()
alld = [data0, data1, data2, data3, data4]
```
