```python
import pandas as pd
import numpy as np
```

```python
#data set mined from openstreetmap data
df = pd.read_csv("Village Coordinates.csv")
```

```python
df
```

```python
Coor = df[['name', 'latitude', 'longitude']].copy()
Coor
```

```python
Coor.rename(columns = {'name':'ward'}, inplace = True)
Coor
```

```python
Coor['ward'] = Coor['ward'].str.upper()
Coor
```

```python
df1 = pd.read_csv("Coordinates.csv")
```

```python
df1
```

```python
inner_join = pd.merge(df1, 
                      Coor, 
                      on ='ward', 
                      how ='inner')
inner_join
```

```python
L = np.setdiff1d(df1.ward, inner_join['ward']).tolist()
L
```

```python
f = inner_join.duplicated(subset=['ward'])
inner_join['TF'] = f
f
```

```python
inner_join
```

```python
t = inner_join[['ward', 'Occurances', 'latitude_y', 'longitude_y']].copy()
t
```

```python
t['Prob'] = t['Occurances']/t['Occurances'].sum()
t
```

```python
t.to_csv('Proba.csv')  
```
