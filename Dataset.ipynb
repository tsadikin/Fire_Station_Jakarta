```python
 import pandas as pd
```

```python
df = pd.read_csv("Data_Kebakaran_Compiled.csv")
df.drop(df.index[8762:8807], axis=0, inplace=True)
```

```python
df
```

```python
df = df.rename(columns={'jenis_kejadian_bencana': 'Disaster_Type', 'kelurahan': 'ward', 'kecamatan': 'district', 'wilayah': 'region', 'taksiran_kerugian': 'Total_Loss'})
df
```

```python
replace_values = {'Penyelamatan' : 'Rescue', 'Kebakaran' : 'Fire', 'Bantuan Operasi' : 'Assistance', 'Bencana Kebakaran' : 'Fire', 'Bencana pada Ketinggian (operasi penyelamatan Beda Ketinggian)' : 'Rescue', 'Pertolongan Gawat Darurat (Ambulan Transportasi).' : 'Rescue', 'Bencana Bangunan Runtuh (operasi penyelamatan Collapse Structure).' : 'Rescue', 'Bencana pada Air (operasi penyelamatan pada Air)': 'Rescue', 'Pertolongan Gawat Darurat (Ambulan Paramedik)': 'Rescue', 'Bencana Transportasi (operasi penyelamatan Transportasi).' : 'Fire'}
df1 = df.replace({"Disaster_Type": replace_values})
df1
```

```python
df2 = df1[df1["Disaster_Type"].str.contains("Bencana penanganan B3 (operasi penyelamatan B3)") == False]
df2
```

```python
df2['ward'] = df2['ward'].str.upper()
```

```python
df3 = pd.DataFrame(df2.ward.value_counts().reset_index().values, columns=["ward", "Occurances"])
df3index = df3.sort_index(axis = 0, ascending=True)
df3index
```

```python
df3index.to_csv('Ward_Occurances.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
plt.bar(df3index['ward'],df3index['Occurances'])
plt.title('Occurances')
plt.xlabel('ward')
plt.ylabel('Occurances')
plt.show()
```

```python
df3index['Occurances'].sum()
```

```python
df3index['Probability'] = df3index['Occurances']/df3index['Occurances'].sum()
df3index
```
